```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime


```

```python
# Cargar el archivo CSV
df = pd.read_csv('/content/drive/MyDrive/Practico Coink/Data Proyec Coink/depositos_oinks.csv')
```

```python
# Mostrar primeras lineas del DataFrame
df.head()
```

# Transformar Data-Set

```python
# Eliminar la columna "Unnamed: 0"
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Convertir la columna "operation_value" a tipo int
df['operation_value'] = df['operation_value'].astype(int)
```

```python
# Convertir las columnas "operation_date" y "user_createddate" a tipo fecha
df['operation_date'] = pd.to_datetime(df['operation_date'])
df['user_createddate'] = pd.to_datetime(df['user_createddate'])
```

```python
# Crear una nueva columna 'operation_hour' con la hora extraída
df['operation_hour'] = df['operation_date'].dt.time

# Eliminar la hora de la columna 'operation_date'
df['operation_date'] = df['operation_date'].dt.date
```

```python
# Crear una nueva columna 'user_created_hour' con la hora extraída
df['user_created_hour'] = df['user_createddate'].dt.time

# Eliminar la hora de la columna 'ser_createddate'
df['user_createddate'] = df['user_createddate'].dt.date


```

```python
# Formatear la columna 'user_created_hour' al nuevo formato
df['user_created_hour'] = df['user_created_hour'].apply(lambda x: x.strftime('%H:%M:%S'))

```

```python
# Convertir la columna 'operation_date' a tipo fecha
df['operation_date'] = pd.to_datetime(df['operation_date'])

# Convertir la columna 'user_createddate' a tipo fecha
df['user_createddate'] = pd.to_datetime(df['user_createddate'])

```

```python
# Convertir la columna 'operation_hour' al tipo hora
df['operation_hour'] = pd.to_datetime(df['operation_hour'], format='%H:%M:%S').dt.time

# Convertir la columna 'user_created_hour' al tipo hora
df['user_created_hour'] = pd.to_datetime(df['user_created_hour'], format='%H:%M:%S').dt.time


```
